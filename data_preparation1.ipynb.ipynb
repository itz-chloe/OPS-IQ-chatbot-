```python

import pandas as pd

df = pd.read_csv('data/DataCoSupplyChainDataset.csv', encoding='unicode_escape')
```

```python
df= df.drop(columns=[
    'Product Description',
    'Order Zipcode',
    'Customer Password'
])

print (df.shape) 
```

```python
df.columns = (df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace(r'[^a-z0-9_]', '', regex=True))

print(df.columns.tolist())
```

```python
orders = df[[
    'order_id', 'order_date_dateorders', 'order_status',
    'days_for_shipping_real', 'days_for_shipment_scheduled',
    'late_delivery_risk', 'delivery_status', 'shipping_mode',
    'sales', 'order_profit_per_order'
]].drop_duplicates(subset='order_id')

customers = df[[
    'customer_id', 'customer_fname', 'customer_lname',
    'customer_city', 'customer_country', 'customer_segment'
]].drop_duplicates(subset='customer_id')

products = df[[
    'order_id', 'product_name', 'category_name',
    'department_name', 'product_price',
    'order_item_quantity', 'order_item_total'
]]

print(f"Orders: {orders.shape}")
print(f"Customers: {customers.shape}")
print(f"Products: {products.shape}")
```

```python
import sqlite3
conn = sqlite3.connect('db/supply_chain.db')
orders.to_sql('orders', conn, if_exists='replace', index=False)
customers.to_sql('customers', conn, if_exists='replace', index=False)
products.to_sql('products', conn, if_exists='replace', index=False)

conn.close()
print("Database created successfully!")
```
